```python
def synthesize_images(orig_img, flow_img, f_of_f_img):
    
    fig = plt.figure(figsize = (30,21))
    rows = 6
    columns = 5
    
    print(orig_img.shape, flow_img.shape, f_of_f_img.shape)
    
    prev = orig_img
    
    for i in range(30):
        
        
        synth_img = prev + (i * flow_img) 
        
        fig.add_subplot(rows, columns, i+1)
        plt.imshow(synth_img)
        plt.axis("off")
        plt.title(" Sythetic Image #%d " % i)
        
        prev = synth_img
    
    
```

```python
def compute_flow_of_flow(path):
    
    fig = plt.figure(figsize = (20, 14))
    rows = 11
    columns = 5
    
    x0 = []
    v0 = []
    a0 = []
    
    cap = cv2.VideoCapture('./circle_noise1.avi')
    
    for i in range(1, 11):
        _, orig = cap.read()
        orig_gray = cv2.cvtColor(orig, cv2.COLOR_BGR2GRAY)
        
        #Fetching first flow
        prev_flow = cv2.imread( path + 'flow1_image_%d.jpg' % i)
        
        #Turning previous flow into gray scale
        prev = cv2.cvtColor(prev_flow, cv2.COLOR_BGR2GRAY)
        
        
        # Fetching second flow and converting to gray scale
        current_flow = cv2.imread(path + 'flow1_image_%d.jpg' % (i+1))
        next = cv2.cvtColor(current_flow, cv2.COLOR_BGR2GRAY)
        
        #Computing flow of flow
        flow = cv2.calcOpticalFlowFarneback(prev, next, None, 0.5, 3, 15, 3, 5, 1.2, 0)
        
        flow_1 = flow[:, :, 0]
        flow_2 = flow[:, :, 1]
        
#         flow_1 = cv2.cvtColor(flow_1, cv2.COLOR_GRAY2BGR)
        cv2.imwrite('./Flows/Flow of flow 1/flow_of_flow1_%d_to_%d.jpg' % (i ,(i+1)), flow_1)
#         cv2.imwrite('./Flows/Flow of flow 2/flow_of_flow2_%d_to_%d.jpg' % (i ,(i+1)), flow_2)
        
        if i == 1:
            fig.add_subplot(rows, columns, (columns * i) + 1)
            plt.imshow(orig_gray)
            plt.axis("off")
            plt.title("Original")

            fig.add_subplot(rows, columns, (columns * i) + 2)
            plt.imshow(prev)
            plt.axis("off")
            plt.title("Prev Flow")

            fig.add_subplot(rows, columns, (columns * i) + 3)
            plt.imshow(flow_1)
            plt.axis("off")
            plt.title("Flow of Flow")

            fig.add_subplot(rows, columns, (columns * i) + 4)
            plt.imshow(next)
            plt.axis("off")
            plt.title("Current Flow") 
            
            x0 = orig_gray
            v0 = prev
            a0 = flow_1
            synthesize_images(x0, v0, a0)
               

#             fig.add_subplot(rows, columns, (columns * i) + 5)
#             plt.imshow(synthetic)
#             plt.axis("off")
#             plt.title("Synthesized Image")

        prev = next
    
        
```
